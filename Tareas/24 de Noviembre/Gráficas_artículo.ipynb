```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("datos_template.xlsx")

# Promediar réplicas
df_avg = df.groupby(["Temperature", "Water activity"]).mean().reset_index()

for T in [25, 35, 45]:
    dft = df_avg[df_avg["Temperature"] == T].copy()
    dft = dft.sort_values("Water activity")

    plt.plot(
        dft["Water activity"],
        dft["Moisture content (% wet basis)"],
        marker='s', linestyle='',
        label=f"{T}°C"
    )

plt.xlabel("aw")
plt.ylabel("Moisture content (% wet basis)")
plt.ylim(0, 20)  # Ajuste solicitado
plt.grid(True)
plt.legend()
plt.show()
```
