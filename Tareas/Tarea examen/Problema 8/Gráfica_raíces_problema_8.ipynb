```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
V = 1e6       # m^3
Q = 1e5       # m^3/año
W = 1e6       # g/año
k = 0.25      # m^0.5/año

f = lambda c: W - Q*c - k*V*np.sqrt(c)

# Rango de 0 a 10
c = np.linspace(0, 10, 500)
y = f(c)

plt.plot(c, y, label='f(c)')
plt.axhline(0, color='red', alpha=0.5)
plt.xlabel('Concentración c [g/m^3]')
plt.ylabel('f(c)')
plt.title('Búsqueda de raíz para concentración en estado estable')
plt.legend()
plt.grid(True)
plt.show()
```
