```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos la función
R=3
Vdeseado=30
fc = lambda h: (np.pi/3)*h**2*(3*R - h) - Vdeseado

# Rango de 0 a 6
h = np.linspace(0,10, 200)  # uso más puntos para suavidad
y = fc(h)

# Graficar
plt.plot(h, y, label=r"$f(h) = \frac{\pi}{3}h^2(3R-h)-30$")
plt.axhline(0, color="gray", alpha=0.5)
plt.xlabel("h")
plt.ylabel("f(h)")
plt.legend()
plt.grid(True)
plt.show()
```
