```python
import numpy as np
from scipy.optimize import minimize

V0 = 0.8
t = 0.03
dens = 8000
Lmax = 2
Dmax = 1
cm = 4.5
cw = 20

def volumen(vars):
    D, L = vars
    return np.pi * D**2 * L / 4

def costo(vars):
    D, L = vars
    vcoraza = (L*np.pi) * (((D/2 + t)**2) - (D/2)**2)
    vtapas  = (2*np.pi) * ((D/2 + t)**2) * t
    m       = (vcoraza + vtapas) * dens
    lw      = (4*np.pi) * (D + t)

    C = cm * m + cw * lw
    return C
cons = ({
    'type': 'eq',
    'fun': lambda vars: volumen(vars) - V0
})
bounds = [(0, 1), (0, 2)]

x0 = [0.5, 1.0]

res = minimize(costo, x0, method="SLSQP", bounds=bounds, constraints=cons)

print("Resultado:")
print(res)
print("\nD óptimo:", res.x[0])
print("L óptimo:", res.x[1])
print("Costo mínimo:", costo(res.x))
```
