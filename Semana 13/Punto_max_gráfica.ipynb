```python
import numpy as np
import matplotlib.pyplot as plt

fc = lambda x: (-0.5*x**2)+(2.5*x)+4.5

x = np.linspace(-10,20, 200)
y = fc(x)

# Graficar
plt.plot(x, y, label="f(x) = (-0.5*x**2)+(2.5*x)+4.5")
plt.axhline(0, color="gray", alpha=0.5)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.ylim(0, 10)
plt.show()
```
